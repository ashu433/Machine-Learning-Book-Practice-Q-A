```python
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import  MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import cross_validate
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.neural_network import MLPRegressor
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
housing_data = pd.read_csv("housing.csv")
housing_data.head()
```

```python
housing_data.info()
```

```python
data = pd.get_dummies(housing_data['ocean_proximity']).astype(int)
data.head()
```

```python
housing_data=housing_data.drop('ocean_proximity',axis=1)
```

```python
housing_data=pd.concat([housing_data,data],axis=1)
```

```python
housing_data.head()
```

```python
housing_data.isnull().sum()
```

```python
impt=SimpleImputer()
imputed_data=impt.fit_transform(housing_data)
imputed_data=pd.DataFrame(imputed_data,columns=housing_data.columns)

imputed_data.isnull().sum()
```

```python
min_max_scalar=MinMaxScaler()
scaled_data=min_max_scalar.fit_transform(imputed_data)
scaled_data=pd.DataFrame(scaled_data,columns=imputed_data.columns)
scaled_data.head()
```

```python
X=scaled_data.drop("median_house_value",axis=1)
Y=scaled_data['median_house_value']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=40)
```

```python
linear_reg=LinearRegression()
linear_model=linear_reg.fit(X_train,Y_train)
Y_predict=linear_model.predict(X_test)
```

```python
print('MAE',mean_absolute_error(Y_test, Y_predict))
print('MSE', mean_squared_error(Y_test, Y_predict))
print('R2 score', r2_score(Y_test, Y_predict))
```

```python
nn = MLPRegressor(random_state=1, max_iter=300,  n_iter_no_change=10)
nn_model = nn.fit(X_train, Y_train)
nn_predictions = nn_model.predict(X_test)
print('mean accuracy on the given test data and labels.', nn_model.score(X_test, Y_test))
```
