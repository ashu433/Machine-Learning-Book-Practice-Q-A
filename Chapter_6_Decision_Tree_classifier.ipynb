```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# **Decision Tree Classifier**

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
iris=load_iris()
iris.keys()
```

```python
X=iris.data[:,2:]
Y=iris.target
```

```python
dec_clf=DecisionTreeClassifier(max_depth=3)
dec_clf.fit(X,Y)
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12, 8))  # Optional, to adjust plot size
plot_tree(dec_clf, filled=True, feature_names=iris.feature_names[2:], class_names=iris.target_names)  # Optional: customize class names
plt.show()
```

```python
species = iris.target_names

# Plot
plt.figure(figsize=(8, 6))
for i in range(3):
    plt.scatter(X[Y == i, 0], X[Y == i, 1], label=species[i])

plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Iris Petal Length vs. Width')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

Z = dec_clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plotting the decision boundary
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.RdYlBu)

# Plot the actual data points
for i in range(3):
    plt.scatter(X[Y == i, 0], X[Y == i, 1], label=iris.target_names[i])

plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Decision Tree Decision Boundary")
plt.legend()
plt.grid(True)
plt.show()
```
